# Scatter plot function:
Plot_Scatter(data,target,exclude,hue,palette,title,no_cols) is a standard function for ploting scatter sub plots in which the points will be grouped by a particular feature and the x-axis will be constant for every graph. And, in each graph the y-axis will be a different feature.
 
 It will take 7 parameters :
 1) df = Is the pandas dataframe in the data is stores
 2) target = Is the target feature(column) for x-axis
 3) exclude = the columns you want to exclude
 4) hue = Is the feature(column) for which you want to group the scatter points
 5) palette = palette combination
 6) title = Title of the plot
 7) no_cols = no.of subgraphs in the graph by default is 3

```python
def Plot_Scatter(data = None,target=None,exclude=None,hue=None,palette=None,title = None,no_cols = 3):
    import math
    if(exclude != None):
        new_col = [col for col in data.columns if col not in exclude]
    else:
        new_col = data.columns
        
    rows = math.ceil(len(new_col)/no_cols)

    fig, axes = plt.subplots(rows, no_cols, figsize=(18, 21))
    if(title == None):
        fig.suptitle('Scatter Sub Plot')
    else:
        fig.suptitle(f'{title}')
    i,j = 0,0    # i,j are for iterating for the sub graphs
    for cols in new_col:
        sns.scatterplot(ax = axes[i,j],data=data,x=target,y= cols,hue=hue,palette=palette)
        j+=1
        if(j >= no_cols):
            j = 0
            i += 1

Plot_Scatter(data=new_df,target='LoanOnCard',exclude=['ID'],title='Scatter Plot for the DataSet Bank')
```

# Pie plot function:
Plot_pie(data,exclude,palette,title,no_cols) is a standard function for ploting pie sub plots.
 
 It will take 5 parameters :
 1) df = Is the pandas dataframe in the data is stores
 2) exclude = the columns you want to exclude
 3) palette = palette combination
 4) title = Title of the plot
 5) no_cols = no.of subgraphs in the graph by default is 3

```python
def plot_pie(data = None,exclude=None,palette=None,title = None,no_cols = 3):
    import math
    if(exclude == True):
        dtype_col = data.select_dtypes(include=['float', 'int']).columns
        new_col = [col for col in data.columns if col not in dtype_col]
    else:
        new_col = data.columns
    
    rows = math.ceil(len(new_col)/no_cols)

    fig, axes = plt.subplots(rows, no_cols, figsize=(18, 10))
    if(title == None):
        fig.suptitle('Pie Sub Plot')
    else:
        fig.suptitle(f'{title}')
    i,j = 0,0    # i,j are for iterating for the sub graphs
    for cols in new_col:
        category_counts = data[cols].value_counts()
        ax = axes[i,j]
        ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
        ax.set_title(f'{cols}')
        j+=1
        if(j >= no_cols):
            j = 0
            i += 1

plot_pie(data=new_df,exclude=True,title='Pie Plot for the DataSet Bank',no_cols = 3)
```

# Hist plot function:
Plot_Hist(data,exclude,hue,color,title,no_cols,kde) is a standard function for ploting hist sub plots in which the points will be grouped by a particular feature. 
 
 It will take 7 parameters :
 1) df = Is the pandas dataframe in the data is stores
 2) exclude = the columns you want to exclude
 3) hue = Is the feature(column) for which you want to group the scatter points
 4) color = color for hist plot
 5) title = Title of the plot
 6) no_cols = no.of subgraphs in the graph by default is 3
 7) kde = by default is False

```python
def Plot_Hist(data = None,exclude=None,hue=None,color='blue',title = None,no_cols = 3,kde=False):
    import math
    if(exclude != None):
        new_col = [col for col in data.columns if col not in exclude]
    else:
        new_col = data.columns
        
    rows = math.ceil(len(new_col)/no_cols)

    fig, axes = plt.subplots(rows, no_cols, figsize=(18, 21))
    if(title == None):
        fig.suptitle('Scatter Sub Plot')
    else:
        fig.suptitle(f'{title}')
    i,j = 0,0    # i,j are for iterating for the sub graphs
    for cols in new_col:
        sns.histplot(ax = axes[i,j], data=data ,x= cols , hue = hue,kde=kde,palette=palette,color=color)
        j+=1
        if(j >= no_cols):
            j = 0
            i += 1

Plot_Hist(data=new_df,exclude=['ID'],kde=True,title='Histogram Plot for the DataSet Bank',color='red')
```
