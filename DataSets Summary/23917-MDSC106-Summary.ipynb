Pandas is a Python library used for working with data sets to manipulate, analyze, clean and explore the data associated with the dataset.

Pandas DataFrame is 2-d tabular labeled data structure, where the data is stored in rows and column order.

Using pandas a large dataset can be analyzed in a easy way.

Pandas has many useful function to handle the datasets in case of missing values, data duplication, or cleaning irrelevant data associated with the dataset for better understanding/output of the data.

Pandas DataFrame can be created usinng python dictionaries, list or .csv files.

While creating a DataFrame in pandas, the no.of elements in the row must be equal to no.of columns.

```python
import seaborn as sns
def diagnostic_plots(df, variable, target):
    plt.figure(figsize=(20,7))
    
    plt.subplot(1,4,1)
    sns.scatterplot(x=df[variable],y=df[target],color ='g')
    plt.title('Scatter')
    
    plt.subplot(1,4,2)
    sns.histplot(df[variable], kde = True, color='r')
    plt.title('Histogram')
    
    plt.subplot(1,4,3)
    sns.barplot(x = target, y = variable, data = df)
    plt.title('Bargraph')
    
    plt.subplot(1, 4, 4)
    sns.boxplot(y=df[target],color = 'b')
    plt.title('Boxplot')
    
    plt.show()
      
```

```python
for col in df:
    if(col != 'sex'):
        diagnostic_plots(df,col,'age')    # For some col the graph is not possible because of null values like sex,bp,s2... .
```

## Scatter SubGraphs For Wine Dataset

```python
# Sub_graphs(DataFrame,Groupby,x_axis,length,rows,cols) is a standard function for ploting scatter sub plots in which the points will be grouped by a particular feature and the x-axis will be constant for every graph. And, in each graph the y-axis will be a different feature
def Sub_graphs(wine_df,Groupby,x_axis,length,rows,cols):
    fig, axes = plt.subplots(rows, cols, figsize=(18, 21))
    fig.suptitle('Scatter Sub Plots Grouped by Class Features')

    temp,i,j = 0,0,0      # temp variable is for iterating over column index and i,j are for iterating for subgraph index
    while (temp < length):
        if ((temp != Groupby) and (temp != x_axis)):
            sns.scatterplot(ax = axes[i,j],data =wine_df, x = wine_df.columns[x_axis] ,y= wine_df.columns[temp], hue = wine_df.columns[Groupby], palette='deep')
            j+=1
            if(j > 2):
                j = 0
                i += 1
        temp += 1

Sub_graphs(wine_df,0,1,14,4,3)  #Sub_graphs(DataFrame,Groupby,x_axis,length,rows,cols)
```

```python
def BoxSub_graphs(wine_df,Groupby,x_axis,length,rows,cols):
    fig, axes = plt.subplots(rows, cols, figsize=(18, 21))
    fig.suptitle('Box Sub Plots Grouped by Class Features')

    temp,i,j = 0,0,0      # temp variable is for iterating over column index and i,j are for iterating for subgraph index
    while (temp < length):
        if ((temp != Groupby) and (temp != x_axis)):
            sns.boxplot(ax = axes[i,j],data=wine_df, x = wine_df.columns[x_axis] ,y= wine_df.columns[temp], hue = wine_df.columns[Groupby], palette='deep')
            j+=1
            
            if(j > 2):
                j = 0
                i += 1
        temp += 1

BoxSub_graphs(wine_df,0,0,13,4,3)    #BoxSub_graphs(df,Groupby,x_axis,length,rows,cols)
plt.show()
```
